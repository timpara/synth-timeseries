# Synthetic Data Generator

This notebook demonstrates how to use `synth-timeseries` to generate and decompose seasonal time series data.

```python
from synth_timeseries.data.generator.cyclical import SeasonalDataGenerator
import matplotlib.pyplot as plt



seasonal_data_generator = SeasonalDataGenerator()
additive_decomposition = seasonal_data_generator.generate_seasonal_additive_series()
multiplicative_decomposition = seasonal_data_generator.generate_seasonal_multiplicative_series
 
# Plotting the decomposed components
# Additive Decomposition
plt.figure(figsize=(14, 8))
additive_decomposition.plot().suptitle('Additive Decomposition', fontsize=22)
plt.tight_layout()
plt.show()

# # Multiplicative Decomposition
plt.figure(figsize=(14, 8))
multiplicative_decomposition.plot().suptitle('Multiplicative Decomposition', fontsize=22)
plt.tight_layout()
plt.show()
```
