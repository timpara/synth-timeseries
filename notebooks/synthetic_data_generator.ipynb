# Synthetic Time Series Generators

This notebook demonstrates all seven generators provided by **synth-timeseries**.

Each generator is shown with default and custom parameters, and the output is plotted using matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
plt.rcParams["figure.figsize"] = (12, 4)
```

## 1. Geometric Brownian Motion (GBM)

Models asset prices with constant drift and volatility. Widely used in finance (e.g. Black-Scholes).

```python
from synth_timeseries import GeometricBrownianMotion

gbm = GeometricBrownianMotion(mu=0.1, sigma=0.2, dt=0.01, num_steps=1000, initial_value=100)
prices = gbm.simulate()

plt.figure()
plt.plot(prices)
plt.title("Geometric Brownian Motion")
plt.xlabel("Time step")
plt.ylabel("Price")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f"Initial: {prices[0]:.2f}, Final: {prices[-1]:.2f}, Min: {prices.min():.2f}, Max: {prices.max():.2f}")
```

## 2. ARIMA Process

Autoregressive Integrated Moving Average process. Combines autoregression, differencing, and moving average components.

```python
from synth_timeseries import ARIMAProcess

arima = ARIMAProcess(ar_params=[0.7, -0.2], ma_params=[0.3], d=1, mu=0.0, sigma=1.0, num_steps=500)
series = arima.simulate()

plt.figure()
plt.plot(series)
plt.title("ARIMA(2,1,1) Process")
plt.xlabel("Time step")
plt.ylabel("Value")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f"Length: {len(series)}, Mean: {series.mean():.2f}, Std: {series.std():.2f}")
```

## 3. GARCH Process

Generalized Autoregressive Conditional Heteroskedasticity. Models time-varying volatility clustering.

```python
from synth_timeseries import GARCHProcess

garch = GARCHProcess(omega=0.1, alpha=0.15, beta=0.8, dt=0.01, num_steps=1000)
returns = garch.simulate()

fig, axes = plt.subplots(2, 1, figsize=(12, 6), sharex=True)

axes[0].plot(returns)
axes[0].set_title("GARCH(1,1) Returns")
axes[0].set_ylabel("Return")
axes[0].grid(True, alpha=0.3)

axes[1].plot(returns ** 2, color="orange")
axes[1].set_title("Squared Returns (volatility proxy)")
axes[1].set_xlabel("Time step")
axes[1].set_ylabel("Squared return")
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## 4. Ornstein-Uhlenbeck Process

Mean-reverting stochastic process. Used to model interest rates, commodity prices, and pairs trading signals.

```python
from synth_timeseries import OrnsteinUhlenbeckProcess

ou = OrnsteinUhlenbeckProcess(theta=0.5, mu=5.0, sigma=0.3, dt=0.01, num_steps=1000)
path = ou.simulate()

plt.figure()
plt.plot(path, label="OU path")
plt.axhline(y=5.0, color="red", linestyle="--", alpha=0.7, label="Long-term mean ($\\mu=5$)")
plt.title("Ornstein-Uhlenbeck Process")
plt.xlabel("Time step")
plt.ylabel("Value")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f"Sample mean: {path.mean():.2f} (long-term mean: 5.0)")
```

## 5. Levy Process

Heavy-tailed stochastic process with jumps. Models financial returns that exhibit extreme events.

```python
from synth_timeseries import LevyProcess

levy = LevyProcess(alpha=1.5, beta=0.0, dt=0.01, num_steps=1000)
path = levy.simulate()

plt.figure()
plt.plot(path)
plt.title("Levy Process ($\\alpha=1.5$)")
plt.xlabel("Time step")
plt.ylabel("Cumulative value")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## 6. Sinusoidal Generator

Generates damped sinusoidal time series with randomized period, damping, and phase. Useful for benchmarking time series models on quasi-periodic data.

```python
from synth_timeseries import SinusoidGenerator

gen = SinusoidGenerator(n=200, n_sinusoid_functions=5)
data = gen.generate_data()

plt.figure(figsize=(12, 5))
for i in range(data.shape[0]):
    plt.plot(data[i], alpha=0.7, label=f"Series {i+1}")
plt.title("Damped Sinusoidal Series")
plt.xlabel("Time step")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## 7. Seasonal Data Generator

Generates synthetic seasonal time series with trend, seasonal, cyclic, and noise components. Supports both additive and multiplicative decomposition.

```python
from synth_timeseries import SeasonalDataGenerator

seasonal_gen = SeasonalDataGenerator()

# View the raw combined components
df = seasonal_gen.load_combined_components_dataframe()
print(f"DataFrame shape: {df.shape}")
print(f"Columns: {list(df.columns)}")
df.head()
```

```python
# Plot raw components
fig, axes = plt.subplots(3, 1, figsize=(14, 10), sharex=True)

axes[0].plot(df.index, df["additive"], linewidth=0.5)
axes[0].set_title("Additive combined series")
axes[0].set_ylabel("Value")
axes[0].grid(True, alpha=0.3)

axes[1].plot(df.index, df["multiplicative"], linewidth=0.5, color="orange")
axes[1].set_title("Multiplicative combined series")
axes[1].set_ylabel("Value")
axes[1].grid(True, alpha=0.3)

axes[2].plot(df.index, df["seasonal"], linewidth=0.5, color="green", label="seasonal")
axes[2].plot(df.index, df["cyclic"], linewidth=0.5, color="red", label="cyclic")
axes[2].set_title("Seasonal and cyclic components")
axes[2].set_xlabel("Date")
axes[2].set_ylabel("Value")
axes[2].legend()
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Additive decomposition
additive_decomposition = seasonal_gen.generate_seasonal_additive_series()

plt.figure(figsize=(14, 8))
additive_decomposition.plot().suptitle("Additive Decomposition", fontsize=18)
plt.tight_layout()
plt.show()
```

```python
# Multiplicative decomposition
multiplicative_decomposition = seasonal_gen.generate_seasonal_multiplicative_series()

plt.figure(figsize=(14, 8))
multiplicative_decomposition.plot().suptitle("Multiplicative Decomposition", fontsize=18)
plt.tight_layout()
plt.show()
```
